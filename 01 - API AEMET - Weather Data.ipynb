# Tratado de datos en PANDAS

```python
# Seleccionamos las columnas que queremos
datos_aemet = api[['fecha','prec','tmin','tmed','tmax','velmedia','racha']]
datos_aemet
```

```python
# Realizamos replace de ',' por '.' para poder pasar los datos de las columnas a float
datos_aemet.prec = datos_aemet.prec.str.replace(',','.')
datos_aemet.prec = datos_aemet.prec.str.replace('Ip','0')
datos_aemet.tmin = datos_aemet.tmin.str.replace(',','.')
datos_aemet.tmax = datos_aemet.tmax.str.replace(',','.')
datos_aemet.tmed = datos_aemet.tmed.str.replace(',','.')
datos_aemet.velmedia = datos_aemet.velmedia.str.replace(',','.')
datos_aemet.racha = datos_aemet.racha.str.replace(',','.')
#datos_aemet.fecha = datos_aemet.fecha.str.replace('-','/')
datos_aemet
```

```python
# Transformamos a float las demas columnas
numeric = ['prec','tmin','tmed','tmax','velmedia','racha']
datos_aemet[numeric] = datos_aemet[numeric].apply(pd.to_numeric, downcast='integer')
datos_aemet.info()
```

```python
# Tratado de nans
datos_aemet['velmedia'].fillna((datos_aemet['velmedia'].mean()), inplace=True)
datos_aemet['racha'].fillna((datos_aemet['racha'].mean()), inplace=True)
datos_aemet
```

```python
# Transformamos la fecha a 'datetime64'
datos_aemet['fecha'] = pd.to_datetime(datos_aemet['fecha'])
datos_aemet.info()
```

```python
# Cambiamos el formato de la fecha para poder realizar el merge
datos_aemet.fecha = datos_aemet.fecha.dt.strftime('%d/%m/%Y')
datos_aemet.head(5)
```

```python
# Todo lo superios a 0.1 es lluvia
prec = datos_aemet[datos_aemet['prec']>= 0.1]
prec
```

```python
datos_aemet['Mojado'] = datos_aemet.apply(lambda x: 1 if x['prec'] >= 0.1 else 0, axis=1)
datos_aemet['Seco'] = datos_aemet.apply(lambda x: 0 if x['prec'] > 0.1 else 1, axis=1)
datos_aemet
```

# Carga de dataframe Original y MERGE

```python
# Cargamos el dataset Original
df = pd.read_pickle("accidentalidad_192021_actualizadoAbril22.pickle")
df = df.reset_index()
df.drop("fecha2", axis=1, inplace=True)
df.head(10)
```

```python
# Generamos un merge entre el dataframe de 'df' y 'datos_aemet'
df_original = pd.concat([df,datos_aemet],axis=1)
#df_final = df_final.drop_duplicates(df_final.columns.to_list())
df_original
```

```python
# Verificar campos de estado_meteorológico
df_original['estado_meteorológico'].unique()
# 1. 'Despejado'
# 2. 'nan'
# 3. 'Se desconoce'
# 4. 'Lluvia débil'
# 5. 'Nublado'
# 6. 'LLuvia intensa'
# 7. 'Granizando'
# 8. 'Nevando'
```

# Verificacion del dataset original y los datos extraidos de AEMET

```python
meteo_dummies = pd.get_dummies(df_original["estado_meteorológico"])
#meteo_dummies['fecha'] = df_original['fecha']
meteo_dummies
```

```python
# Generamos un merge entre el dataframe de 'df' y 'datos_aemet'
df_final = pd.concat([df_original,meteo_dummies],axis=1)
#df_final = df_final.drop_duplicates(df_final.columns.to_list())
df_final
```

```python
# Creamos las nuevas columnas que verifican sobre el get_dummies y las nuevas columnas 'Lluvia' y 'Sol' extraidas del AEMET
tqdm.pandas()
df_final['vrf_Despejado'] = df_final.progress_apply(lambda x: 1 if (x['Despejado']==1)&(x['Seco']==1) else 0, axis=1)
df_final['vrf_Lluvioso'] = df_final.progress_apply(lambda x: 1 if (x['LLuvia intensa']==1)&(x['Mojado']==1) else 0, axis=1)
df_final['vrf_Granizando'] = df_final.progress_apply(lambda x: 1 if (x['Granizando']==1)&(x['Mojado']==1) else 0, axis=1)
df_final['vrf_Nevando'] = df_final.progress_apply(lambda x: 1 if (x['Nevando']==1)&(x['Mojado']==1) else 0, axis=1)
df_final['vrf_Lluvia débil'] = df_final.progress_apply(lambda x: 1 if (x['Lluvia débil']==1)&(x['Mojado']==1) else 0, axis=1)
df_final['vrf_Se_desconoce'] = df_final.progress_apply(lambda x: 1 if (x['Se desconoce']==1)&(x['Mojado']==1) else 0, axis=1)
df_final['vrf_Nublado'] = df_final.progress_apply(lambda x: 1 if (x['Nublado']==1)&(x['Mojado']==1) else 0, axis=1)
```
